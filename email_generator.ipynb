```python
import requests 

username = 'dextermadh'

url = f'https://api.github.com/users/{username}/repos'

response = requests.get(url)

if response.status_code == 200: 
    repos = response.json()
    for repo in repos: 
        print(f'{repo['name']} | {repo['html_url']} | {repo['language']}')  
    
```

```python
repo_dict = {}

for repo in repos: 
    repo_dict[repo['name']] = {
        'url': repo['html_url'], 
        'language': repo['language']
    }
repo_dict 
```

```python
import pandas 

df = pandas.DataFrame(repo_dict).T.reset_index() 
df 
```

```python
df = pandas.DataFrame.from_dict(repo_dict, orient='index').reset_index().rename(columns={'index': 'repo_name'})
```

```python
df.isna().sum() 
```

```python
df = df.dropna() 
df 
```

```python
df.to_csv('app/data/repos.csv', index=False) 
```

```python
# chromadb buildip 
import chromadb 

# connect to chroma
client = chromadb.PersistentClient(path='chroma_db')
collection = client.get_or_create_collection(name='github_repos')

# insert into chroma 
for _, row in df.iterrows(): 
    collection.add(
    ids=[row["repo_name"]],
    documents=[f"Repository: {row['repo_name']}, Language: {row['language']}, URL: {row['url']}"],
    metadatas=[{
        "name": row["repo_name"],
        "url": row["url"],
        "language": row["language"]
    }]
)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model='openai/gpt-oss-120b',
    temperature=0.1
)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
from bs4 import BeautifulSoup 
import time

# configure selecnium to run headless 
chrome_options = Options() 
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-dev-shm-usage') 

# start the chrome 
driver = webdriver.Chrome(options=chrome_options)

# open the url 
url = 'https://xpress.jobs/Jobs/View/182000/machine-learning-engineer-internship'
driver.get(url) 

time.sleep(5) # wait until the page loads 

# get the rendered html 
html = driver.page_source 
driver.quit() 

# parse the html with beautifulsoup 
soup = BeautifulSoup(html, 'html.parser')

page_text = soup.get_text(separator='\n', strip=True) 
print(page_text )
```

```python
from langchain_core.prompts import PromptTemplate 

extract_prompt = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_text}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing the 
    following keys: `role`, `experience`, `skills` and `description`.
    Only return the valid JSON. Output Should be valid JSON not a LIST or DICT.
    ### VALID JSON (NO PREAMBLE): 
       
    """ 
)

chain_extract = extract_prompt | llm 
res = chain_extract.invoke(input={'page_text': page_text})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser() 
json_res = json_parser.parse(res.content)
json_res 
```

```python
job = json_res 
job['skills']
```
